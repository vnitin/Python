```python
class Bike:
	
	__total_number_of_bikes = 0
	__rent_per_hour = 5
	__rent_daily = 20
	__rent_per_week = 60

	@classmethod
	def add_bike_to_inventory(cls, number_of_bikes):
		cls.__total_number_of_bikes = number_of_bikes

	@classmethod
	def list_of_available_bikes(cls):
		return cls.__total_number_of_bikes

	@classmethod
	def calculate_rent(cls, number_of_bikes, type_of_rent):
		pass

	@classmethod
	def rent_bike(cls, number_of_bikes, type_of_rent):
		cls.__total_number_of_bikes -= number_of_bikes

	def __init__(self, number_of_bikes):
		Bike.add_bike_to_repository(number_of_bikes)


class Rental_Agent:

	@classmethod
	def issue_bill(cls, customer_obj):
		if customer_obj.number_of_bikes >= 3 and customer_obj.number_of_bikes <= 5:
			if customer_obj.type_of_rent == "hourly":
				return customer_obj.number_of_bikes * cls.__rent_per_hour * 0.3
			elif customer_obj.type_of_rent == "daily":
				return customer_obj.number_of_bikes * cls.__rent_daily * 0.3
			elif customer_obj.type_of_rent == "weekly":
				return customer_obj.number_of_bikes * cls.__rent_per_week * 0.3
		else:
			if customer_obj.type_of_rent == "hourly":
				return customer_obj.number_of_bikes * cls.__rent_per_hour 
			elif customer_obj.type_of_rent == "daily":
				return customer_obj.number_of_bikes * cls.__rent_daily
			elif customer_obj.type_of_rent == "weekly":
				return customer_obj.number_of_bikes * cls.__rent_per_week

	@classmethod
	def issue_bike(cls, number_of_bikes, type_of_rent):
        if number_of_bikes > Bike.list_of_available_bikes():
			return "Sufficient bikes are not available for rent"
		Bike.rent_bike(number_of_bikes, type_of_rent)

	@classmethod
	def display_inventory(cls):
		Bike.list_of_available_bikes()

class Customer:

	def __init__(self, name, number_of_bikes, type_of_rent):
		self.__customer_name = name
		self.__number_of_bikes = number_of_bikes
		self.__type_of_rent = type_of_rent
		Rental_Agent.issue_bike(number_of_bikes, type_of_rent)

	@property
	def number_of_bikes(self):
		return self.__number_of_bikes

	@property
	def type_of_rent(self):
		return	self.__type_of_rent

	def return_bike(self):
		bill_value = Rental_Agent.issue_bill(self)
		return bill_value

	def display_available_bikes(self):
		return Rental_Agent.display_inventory()
```
